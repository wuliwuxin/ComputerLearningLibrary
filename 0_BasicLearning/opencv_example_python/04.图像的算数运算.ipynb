```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
rain = cv.imread("./image/rain.jpg")
```

```python
plt.imshow(rain[:,:,::-1])
```

```python
view = cv.imread("./image/view.jpg")
```

```python
plt.imshow(view[:,:,::-1])
```

```python
img1 = cv.add(rain,view)
```

```python
plt.imshow(img1[:,:,::-1])
```

```python
img2 = rain+view
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
img3 = cv.addWeighted(view,0.3,rain,0.7,0)
```

```python
plt.imshow(img3[:,:,::-1])
```

# 减法

```python
img3 = cv.subtract(view,rain)
```

```python
plt.imshow(img3[:,:,::-1])
```

```python
img4 = view-rain
```

```python
plt.imshow(img4[:,:,::-1])
```
