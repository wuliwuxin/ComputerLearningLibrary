```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("./image/cat.jpeg",0)
```

```python
plt.imshow(img,cmap= plt.cm.gray)
```

# 

```python
hist = cv.calcHist([img],[0],None,[256],[0,256])
```

```python
plt.figure(figsize=(10,8))
plt.plot(hist)
plt.show()
```

# 

```python
plt.imshow(img,cmap=plt.cm.gray)
```

```python
# 
mask = np.zeros(img.shape[:2],np.uint8)
mask[400:650,200:500] = 1
```

```python
plt.imshow(mask,cmap=plt.cm.gray)
```

```python
mask_img = cv.bitwise_and(img,img,mask=mask)
```

```python
plt.imshow(mask_img,cmap=plt.cm.gray)
```

```python
mask_hist = cv.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.plot(mask_hist)
plt.show()
```

```python
img = cv.imread("./image/horse.jpg",0)
```

```python
x = cv.Sobel(img, cv.CV_8U, 1, 0)
y = cv.Sobel(img, cv.CV_8U, 0, 1)
```

```python
Scale_absX = cv.convertScaleAbs(x)  # convert   scale 
Scale_absY = cv.convertScaleAbs(y)
```

```python
result = cv.addWeighted(Scale_absX, 0.5, Scale_absY, 0.5, 0)
```

```python
plt.imshow(result,cmap=plt.cm.binary)
```
